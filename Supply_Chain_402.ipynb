```python
def date_time(df, day_col, month_col, year_col, col_name):
    temp_df = df[[year_col, month_col, day_col]].rename(columns={year_col: 'year', month_col: 'month', day_col: 'day'})
    df[col_name] = pd.to_datetime(temp_df)
   
```

```python
import pandas as pd

# Assuming orders_df and inventory_df have been loaded appropriately

# Convert 'Order YearMonth' to a uniform string format if necessary
orders_ship['Order YearMonth'] = orders_ship['Order YearMonth'].astype(str)

# Aggregating order quantities by 'Product Name' and 'Order YearMonth'
order_agg = orders_ship.groupby(['Product Name', 'Order YearMonth'])['Order Quantity'].sum().reset_index()
```

```python
# Ensure 'Year Month' in inventory_df is in the same format as in orders_df
inv['Year Month'] = inv['Year Month'].astype(str)

# Merging aggregated order data with inventory data
merged_df = pd.merge(order_agg, inv, left_on=['Product Name', 'Order YearMonth'], right_on=['Product Name', 'Year Month'], how='outer')

# Fill missing values which might indicate no inventory or no orders with zeros
merged_df.fillna(0, inplace=True)

```

```python

merged_df['Inventory to Sales Delta'] = merged_df['Warehouse Inventory'] - merged_df['Order Quantity']
```
